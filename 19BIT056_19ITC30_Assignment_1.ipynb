```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

titanic = pd.read_csv('/content/sample_data/train.csv')
```

```python
scatterplot = titanic.plot.scatter('Age', 'Fare')
plt.title("Scatter Plot")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show(scatterplot)
```

```python
plt.scatter(titanic['Age'], titanic['Fare'], c=titanic['Pclass'],
            s=titanic['Pclass'])
plt.title("Scatter Plot")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.colorbar()
plt.legend()
plt.show()
```

```python

titan=titanic.head(20)
plt.plot(titan['Age'],marker='o')
plt.plot(titan['Fare'],marker='o')
plt.title("Line Chart")
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.show()
```

```python
plt.hist(titanic['Age'], color = "skyblue", ec="blue")
plt.title("Histogram")
plt.show()
```

```python
sns.boxplot(x='Age', y="Sex",data=titanic );
plt.show()
```

```python
pd.crosstab(titanic['Pclass'],titanic['Sex']).plot(kind='bar', stacked='True')
plt.title('Stacked Bar Chart')
```
